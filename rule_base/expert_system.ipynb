```python
class HeartExpert(KnowledgeEngine):

    @Rule(Fact(age=P(lambda x:x > 50))&
          Fact(chol=P(lambda x:x >240)))
    def high_risk(self):
        print("High Risk: Multiple critical conditions")

    @Rule(Fact(age=P(lambda x:x > 50)))
    def age_risk(self):
        print("Warning: Age over 50 increases heart disease risk!")
    
    @Rule(Fact(exang=1))
    def angina(self):
        print("Risk: Exercise-induced angina")
    
    @Rule(Fact(age=P(lambda x:x < 40))&
          Fact(chol=P(lambda x:x <200))&
          Fact(trestbps=P(lambda x:x <120)))
    def low_risk(self):
        print("Low Risk: Healthy condition")

    @Rule(Fact(ca=P(lambda x: x > 1)))
    def vessels(self):
        print("Risk: Blocked vessels")
    
    @Rule(Fact(thalach=P(lambda x: x < 120)))
    def low_hr(self):
        print("Risk: Low max heart rate")

    @Rule(Fact(oldpeak=P(lambda x:x >2)))
    def oldpeak_risk(self):
        print("Risk: ST depression")

    @Rule(Fact(age=P(lambda x: x > 50)) &
        Fact(trestbps=P(lambda x: x >= 140)))
    def age_bp(self):
        print("High Risk: Age + High BP")

    @Rule(Fact(cp=P(lambda x: x >= 2)))
    def chest_pain(self):
        print("Risk: Significant chest pain detected")

    @Rule(Fact(fbs=1))
    def sugar(self):
        print("Risk: High fasting blood sugar")
    
    @Rule(Fact(exang=1) &
    Fact(oldpeak=P(lambda x: x > 2)) &
    Fact(thalach=P(lambda x: x < 120)))
    def heart_attack_risk(self):
        print("CRITICAL: High chance of heart disease / heart attack")
   
```

```python
engine = HeartExpert()
engine.reset()

print("=== Welcome to the Heart Disease Expert System ===")
print("Please enter the patient's data as requested:\n")

try:
    patient_info = {
        "age": float(input("Enter age: ")),
        "chol": float(input("Enter cholesterol level : ")),
        "trestbps": float(input("Enter resting blood pressure : ")),
        "exang": float(input("Exercise induced angina? (1 for Yes, 0 for No): ")),
        "thalach": float(input("Enter maximum heart rate achieved : ")),
        "oldpeak": float(input("Enter ST depression induced by exercise : ")),
        "ca": float(input("Number of major vessels colored by flourosopy (0-3): ")),
        "cp": float(input("Chest pain type (0-3): ")),
        "fbs": float(input("Fasting blood sugar > 120 ? (1 for Yes, 0 for No): "))
    }
    
    for key, value in patient_info.items():
        engine.declare(Fact(**{key: value}))

    print("\n--- Expert System Diagnosis Results ---\n")
    
    engine.run()

except ValueError:
    print("\n[Error]: Invalid input! Please enter numbers only.")
```
